```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
pokemons = pd.read_csv('Pokemon.csv')
```

```python
print(pokemons.columns)
```

```python
data = dict(pokemons['Type 1'].value_counts())
x = list(data.keys())
y = list(data.values())

newData = pd.DataFrame({'Type': x, 'Number of pokemons': y})

plt.figure(figsize = (12, 8))

sns.barplot(data = newData, x = 'Type', y='Number of pokemons')

plt.title("Number of pokemons in all generations by their first types")

plt.show()
```

```python
data = dict(pokemons['Type 2'].value_counts())
x = list(data.keys())
y = list(data.values())

newData = pd.DataFrame({'Type': x, 'Number of pokemons': y})

plt.figure(figsize = (12, 8))

sns.barplot(data = newData, x = 'Type', y='Number of pokemons')

plt.title("Number of pokemons in all generations by their secundary types")

plt.show()
```

```python
pokemonsStrongestTypes = pokemons[pokemons['Total'] > 650]


plt.figure(figsize = (10, 8))
sns.countplot(x = 'Type 1', hue = pokemonsStrongestTypes['Type 1'], data = pokemonsStrongestTypes)
plt.title('Counts by Rating for types of Pokemon with total power bigger than 650')
plt.ylabel('Number of pokemons')
plt.xlabel('Type of pokemon')
plt.show()  
```

```python
pokemonsTypesSP_ATK = pokemons[pokemons['Sp. Atk'] > 140]

plt.figure(figsize = (10, 8))
sns.countplot(x = 'Type 1', hue = pokemonsTypesSP_ATK['Type 1'], data = pokemonsTypesSP_ATK)
plt.title('Counts by Rating for types of Pokemon with SP ATK bigger than 140')
plt.ylabel('Number of pokemons')
plt.xlabel('Type of pokemon')
plt.show() 
```

```python
pokemonsTypesATK = pokemons[pokemons['Attack'] > 140]

plt.figure(figsize = (10, 8))
sns.countplot(x = 'Type 1', hue = pokemonsTypesATK['Type 1'], data = pokemonsTypesATK)
plt.title('Counts by Rating for types of Pokemon with ATK bigger than 140')
plt.ylabel('Number of pokemons')
plt.xlabel('Type of pokemon')
plt.show() 
```

```python
data = dict(pokemons['Generation'].value_counts())
x = list(data.keys())
y = list(data.values())


plt.figure(figsize = (12, 8), facecolor='white')
newData = pd.DataFrame({'Generation': x, 'Number of pokemons': y})


palette_color = sns.color_palette('bright') 

plt.pie(data = newData, labels='Generation', x='Number of pokemons', colors= palette_color,autopct='%.0f%%')

plt.title("Percentage of pokemons by generation")


plt.show()
```
